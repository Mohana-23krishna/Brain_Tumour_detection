```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
import random

# Load trained classification model
model = YOLO('runs/classify/Chest_XRay_Datasets5/weights/best.pt')

# Test images directory
test_images_folder = 'Chest_XRay_Datasets/test'

# Get class folders (NORMAL, PNEUMONIA, etc.)
class_folders = [f for f in os.listdir(test_images_folder) 
                 if os.path.isdir(os.path.join(test_images_folder, f))]

# Plot settings
fig, ax = plt.subplots(4, 4, figsize=(16, 16))
ax = ax.ravel()

for idx in range(16):

    # Random class folder
    img_folder = random.choice(class_folders)

    # Get images inside that folder
    img_folder_path = os.path.join(test_images_folder, img_folder)
    img_files = [f for f in os.listdir(img_folder_path) 
                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Random image
    img_path = os.path.join(img_folder_path, random.choice(img_files))

    image = cv2.imread(img_path)

    # Perform classification
    results = model(img_path)
    result = results[0]

    # Prediction
    class_id = int(result.probs.top1)
    class_name = model.names[class_id]
    confidence = float(result.probs.top1conf)

    # Convert color for matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display
    ax[idx].imshow(image_rgb)
    ax[idx].set_title(
        f"Actual: {img_folder}\nPredicted: {class_name} ({confidence:.2f})",
        fontsize=11
    )
    ax[idx].axis('off')

plt.tight_layout()
plt.show()
```
